```python
import requests
import time

API_KEY = 'METTERE API KEY'

def retrieveEverything(query):
    
    results = []
    
    URL = 'https://api.elsevier.com/content/search/scopus'
    time.sleep(30)

    params = {
        'apiKey': API_KEY,
        'query': query,
        #'count': 10,  # Number of results to retrieve
        'sort': 'relevance'  # Sorting options: 'relevance', 'date'
    }

    # API request
    response = requests.get(URL, params=params)

    # Check if success (200)
    if response.status_code == 200:
        
        data = response.json()

        for entry in data['search-results']['entry']
        
            doi = entry.get('prism:doi', 'DOI not available')

            title = entry.get('dc:title', 'Title not available')
            results.append([title, doi, query])
    else:
        print('Error:', response.status_code)
        
    return results
```

```python
queries = """
"Green AI" AND "governance"
"Sustainable AI" AND "policy"
"Environmental AI" AND "governance and policy"
"Energy-efficient AI" AND "regulation"
"Carbon-neutral AI" AND "technology governance"
"AI ethics" AND "environmental policy"
"Green computing" AND "AI governance"
"AI regulations" AND "sustainability"
"Eco-friendly AI" AND "policy framework"
"AI for environmental sustainability" AND "governance"
"Resource-efficient AI" AND "technology policy"
"Green technology adoption" AND "AI governance"
"AI impact assessment" AND "environmental policy"
"AI and sustainable development" AND "regulations"
"Green technology" AND "AI policy"
"""
lines = queries.strip().split('\n')
# lines
```

```python
a = []
for q in lines:
    r = retrieveEverything(q)
    a.append(r)
    # print(r)
res = []
for el in a:
    for e in el:
        res.append(e)
# res
```

```python
import pandas as pd

df = pd.DataFrame(res, columns=['titolo', 'doi', 'query'])
df
```
